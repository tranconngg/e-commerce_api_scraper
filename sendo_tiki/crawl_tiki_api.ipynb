```python
import json
import csv

with open('data_tiki.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

all_products = []
for item in data:
    name = item.get('name', 'N/A')
    price = item.get('price', 'N/A')
    all_products.append([name, price])

with open('All_products.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Name', 'Price'])  # Tiêu đề cột
    writer.writerows(all_products)

print("Data has been saved to 'All_products.csv'")
```
